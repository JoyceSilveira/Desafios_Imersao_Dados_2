#Solução dos desafios da aula 03
####Análise de uma amostra de dados do Enem 2019

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

dados = "https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

df = pd.read_csv(dados)
df.head()
```

#Desafio 1
###Plotar as médias, medianas e moda nas notas de LC e MT


```
_ = plt.figure(figsize=(12,8))
_ = sns.histplot(df, x = "NU_NOTA_LC", kde=True)
_ = plt.axvline(x = df["NU_NOTA_LC"].mean(), color='red')
_ = plt.axvline(x = df["NU_NOTA_LC"].median(), color='orange')
_ = plt.axvline(x = df["NU_NOTA_LC"].mode()[0], color='green')
plt.show()
```

```
_ = plt.figure(figsize=(12,8))
_ = sns.histplot(df, x = "NU_NOTA_MT", kde=True)
_ = plt.axvline(x = df["NU_NOTA_MT"].mean(), color='red')
_ = plt.axvline(x = df["NU_NOTA_MT"].median(), color='orange')
_ = plt.axvline(x = df["NU_NOTA_MT"].mode()[0], color='green')
plt.show()
```

#Desafio 2
###Melhorar a visualização da matriz de correlação e analisar mais detalhadamente.

```
provas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_MT', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']
_ = sns.heatmap(df[provas].corr(), cmap="Blues", annot=True)
plt.show()
```

A nota da prova de ciências humanas é a que mais possui correlação com outras provas (especialmente linguagem e código), com exceção da nota de redação, que por sua vez não possui muita correlação com as demais provas.
